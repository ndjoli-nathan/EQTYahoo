## 2. <u> Basic usage example: </u>
---

#### <u> First, let's suppose we (don't know &) want to know what are the top 5 Tech Companies in the US & in France: </u>

```python
# Top 5 companies in the Technology sector in the United States sorted by Market Capitalization on the NASDAQ:
EQT.Get(country='United States', exchange='NMS', sector='Technology', currency='USD',sorted='cap', length=5)
```

```python
#Top 5 companies in the Technology sector in France sorted by Market Capitalization on the Paris Stock Exchange:
EQT.Get(country='France', exchange='PAR', sector='Technology', currency='EUR',sorted='cap', length=5)
```

#### <u> Now let's suppose we want to fetch 5Y historical data for both the previous queries: </u>

```python
#Fetching Data for the top 5 companies in the Technology sector in the United States sorted by Market Capitalization on the NASDAQ:
USTech5 = EQT.GetData(country='United States', exchange='NMS', sector='Technology', currency='USD',sorted='cap', length=5, start='2020-01-16', end='2025-01-16')
display(USTech5.tail(5))
```

```python
##Fetching Data for the top 5 companies in the Technology sector in France sorted by Market Capitalization on the Paris Stock Exchange:
FRTech5 = EQT.GetData(country='France', exchange='PAR', sector='Technology', currency='EUR',sorted='cap', length=5, start='2020-01-16', end='2025-01-16')
display(FRTech5.tail(5))
```

#### <u>Then you can easily plot the data or process it as you want:</u>

```python
def DrawPlot(df,col):
    matplotlib.pyplot.figure(figsize=(12, 4))
    if col != None:
        for ticker in df[col].columns:
            matplotlib.pyplot.plot(df[col][ticker], label=ticker,linewidth=1)
    else:
        for ticker in df.columns:
            matplotlib.pyplot.plot(df[ticker], label=ticker,linewidth=1)
    matplotlib.pyplot.legend()
    matplotlib.pyplot.show()
```

```python
#Plotting the data for the top 5 companies in the Technology sector in the United States sorted by Market Capitalization on the NASDAQ:
DrawPlot(USTech5,'Close')
```

```python
# Plotting the data for the top 5 companies in the Technology sector in France sorted by Market Capitalization on the Paris Stock Exchange:
DrawPlot(FRTech5,'Close')
```

```python
# Calculate & "visualize" returns.
USTech5Rets = USTech5['Close'].pct_change().dropna()
DrawPlot(USTech5Rets,None)

FRTech5Rets = FRTech5['Close'].pct_change().dropna()
DrawPlot(FRTech5Rets,None)
```
